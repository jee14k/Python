## Initial Value Problem

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def model(y,t):
    dydt = 1
    return dydt

#initial condition
y0 = 0

#time points
t = np.linspace(0,10)
#solve ODE
y = odeint(model,y0,t)

#plot results
plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show()
```

```python
def model(y,t):
    dydt = t
    return dydt

#initial condition
y0 = 0

#time points
t = np.linspace(-5,5)
#solve ODE
y = odeint(model,y0,t)

#plot results
plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show()
```
