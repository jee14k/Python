```python
new_df['[The instructions were clear]'] = pd.concat([data['[The instructions were clear]'], data['[The instructions were clear].1']]).dropna()
```

```python
#new_df.reset_index()
```

```python
new_df.head()
```

```python
new_df['[The event was challenging enough]'] = pd.concat([data['[The event was challenging enough]'], data['[The event was interesting]']]).dropna()
```

```python
new_df['[The event started on time]'] = pd.concat([data['[The event started on time]'], data['[The event started on time].1']]).dropna()
```

```python
new_df['[The volunteers were polite]'] = data['[The volunteers were polite]']
```

```python
new_df['[The venue decoration was suitable/significant]'] = pd.concat([data['[The venue decoration was suitable/significant]'], data['[The venue decoration was relevent]']]).dropna()
```

```python
new_df['[The time duration was sufficient]'] = pd.concat([data['[The time duration was sufficient]'], data['[The time duration was appropriate]']]).dropna()
```

```python
new_df['[The event was professionally managed]'] = pd.concat([data['[The event was professionally managed]'], data['[The event was professionally managed].1']]).dropna()
```

```python
new_df['[The quality of the content was up to your expectations]'] = pd.concat([data['[The quality of the content was up to your expectations]'], data['[The event was entertaining and enjoyable]']]).dropna()
```

```python
new_df['[The necessary supplies were well facilitated]'] = data['[The necessary supplies were well facilitated]']
```

```python
new_df.reset_index(inplace=True)
```

```python
new_df.drop('index', axis=1, inplace=True)
```

```python
new_df.sort_values(by=list(new_df.columns), axis=0, inplace=True)
```

```python
new_df.head()
```

```python
new_df['Fest'] = data['Which fest did you participate in?']
```

```python
for column in list(new_df.columns):
    plt.figure(figsize=(15,8))
    sns.set(style="ticks")
    sns.countplot(x = column, data=new_df, hue='Fest', edgecolor=sns.color_palette("deep", 3))
```

```python
list(new_df.columns)[:9]
```

```python
new_df['Fest'] = new_df['Fest'].apply(lambda x: x.split(' (')[0])
```

```python
for column in list(new_df.columns)[:9]:
    sns.set(style="darkgrid")
    plt.figure(figsize=(15, 8))
    plt.title(column, fontweight="bold", fontsize=20)
    plt.xlabel('FEST', fontweight="bold", fontsize=15)
    sns.countplot(x = 'Fest', hue = column, data=new_df)
```
